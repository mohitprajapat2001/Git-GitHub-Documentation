Now after installation it is important to make some configuration for yourself like adding username, email, name with your git account locally.
So whenever you change a file your configuration can add with the changes. 
Git comes with a tool called git config that lets you get and set configuration variables that control all aspects of how Git looks and operates.

- Command to List all The Available Configuration:  ```$ git config --list --show-origin```
- Get Specific Information: ```$ git config user.name || $ git config user.email || $ git config user.username```

***Sample Output***  




### Update Your Information

The first thing you should do when you install Git is to set your user name and email address. 
This is important because every Git commit uses this information.

```$ git config --global user.name "John Doe"```

```$ git config --global user.email johndoe@example.com```

## Using Local

### The Configuration with Global changes the details for whole system, but We can change the details directory specific

```git config user.name "Your Name Here"```

```git config user.email your@email.example```


***OR***

```git config --local user.name "John Doe"```

```git config --local user.email john@doe.com```
