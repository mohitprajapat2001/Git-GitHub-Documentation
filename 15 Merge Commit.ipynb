#### Merging the Branches - Rebase & Squash Approach

Now First we need to understand what squash actually is ?

To "squash" in Git means to combine multiple commits into one. You can do this at any point in time (by using Git's "Interactive Rebase" feature), though it is most often done when merging branches.

Enough of the reading, lets take some example.
***better to perform all these operations in new branch***



You can see that in this repo i have 4 commits now i wanted to make these 4 into 1.

```git rebase -i HEAD~N```

```git rebase -i COMMIT-SHA```




This will open Editor
***This editor will open in front of you, multiple options are these you are required to choose between then.***



You will see the pick command along with sha-abbrevet and commit message in the first line, this is the place where all things take place.



Mutiple Commands are There: 



1. r, reword is used to select the commit message and update that commit.
2. s, squash is used to remove the commit.
3. p, pick is used as reword but do not give option to update commit message.
4. d, drop is used as drop commit message in updated commit.
5. More Commands are also there.
