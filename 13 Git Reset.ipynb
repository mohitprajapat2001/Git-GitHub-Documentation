Suppose, A scenerio where you wanted to go back in timeline of commit and make some changes there. 
Here comes the role of git reset this allows you to go back in time and reset some changes.

For Examples - As I say unneccessary files dont required to be in direcotry but ***.ipynb_checkpoints*** file is there, lets go back in time and update the .gitignore so that we can change to timeline.



Now we are back in that commit with all the files update, just update the .gitignore remove .ipynb_checkpoints directory from staging area and commit.

After Reset the local branch is backward from remote branch and now we have to push force in remote

```git push origin <branch-name> --force```
