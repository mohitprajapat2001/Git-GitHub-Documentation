You can Technically start with git in two ways,

1. clone: use another git repository to work with it.
2. init: create own git repository from scratch

### Initializing a Repository in an Existing Directory

If you have a project directory that is currently not under version control and you want to start controlling it with Git, you first 
need to go to that project’s directory.

1. #### Empty Repository : git init
- Use cd command to go inside the working directory,

```cd /home/user/mysampledirectory/```

- Create a Empty Git Repository with,

```git init```
This creates a new subdirectory named .git that contains all of your necessary repository files — a Git repository skeleton.

*later we will talk about stagging and commiting all these things just remember ```git init``` to create empty repository.*

2. #### Cloning an Existing Repository

- If you want to get a copy of an existing Git repository — for example, a project you’d like to workon
- ```git clone <repo-url>```
- ```git clone <repo-url> MyCustomRepoName```

When we are working on files, git track those modifications which we can see in each commit which holds the information of file modifications, which we called *snapshot*.

In your working directory can be in one of two states: *tracked* or *untracked*

Tracked files are files that were in the last *snapshot*, as well as any newly staged files; they can be unmodified, modified, or staged. In short, tracked files are files that Git knows about.

### we can check the status of track and untracked files using git command - ```git status```



#### Short Status - ```$ git status -s```



-To track the modified files and add in *stagged area* - ```git add <filename>```




***stagging area - area which track the modified files, stagging area is the area responsible for git to understand the files for snapshot. only the files in stagged area are snapshot and committed***

### Committing Your Changes

Now that your staging area is set up the way you want it, you can commit your changes. Remember that anything that is still unstaged — any files you have created or modified that you haven’t run git add on since you edited them — won’t go into this commit. They will stay as modified files on your disk. 

the last time you ran git status, you saw that everything was staged, so you’re ready to commit your changes. The simplest way to commit is to type ***git commit***

```$ git status```



```$ git add .```
```$ git status```
