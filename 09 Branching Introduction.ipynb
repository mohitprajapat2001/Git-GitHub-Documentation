Nearly every VCS has some form of branching support. Branching means you diverge from the main line of development and continue to do work without messing with that main line. In many VCS tools, this is a somewhat expensive process, often requiring you to create a new copy of your source code directory, which can take a long time for large projects.

Some people refer to Git’s branching model as its “killer feature,” and it certainly sets Git apart in the VCS community. Why is it so special? The way Git branches is incredibly lightweight, making branching operations nearly instantaneous, and switching back and forth between branches generally just as fast. Unlike many other VCSs, Git encourages workflows that branch and merge often, even multiple times in a day. 

*As you may remember from What is Git?, Git doesn’t store data as a series of changesets or differences, but instead as a series of snapshots.*

Enough of Talking about branch, let me explain this with a good examples.

- lets say you are working on a large project with a large no of files and in each file, there is a thousands of line of codes.
- You are working with VCS - git, and comes a long way in same master branch.
- But, now you are at a stage where everything working fine, code is proper, no error, no issue.
- Client Asked to add some sort of feature, now its very risky to work in same flow of master branch.
- Lets say master branch is the branch where your project is live, so if we work in same flow then a great chances are we got some error.

- Here, comes the lifesaver concept of branching, in this case we create a branch named feature.
- We work on this feature branch all the code we write do not affect the master branch.
- and when we are done, we merge the feature branch changes into master branch.
- and later delete that branch. 
- This is the actual workflow used in Companies. So Remember to Understand the concepts on branching

#### Creating a New Branch

What happens when you create a new branch? Well, doing so creates a new pointer for you to move around. Let’s say you want to create a new branch called testing.

```git branch testing```



#### List Branch
