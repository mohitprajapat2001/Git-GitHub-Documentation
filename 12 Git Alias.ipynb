So, You don't love to use shortcuts, they are easy, less typing, easy maintaining. I Definately love shortcuts.

So, why not to create shortcuts in Git, some git commands are to frequent like, 

1. ```add .```
2. ```commit -m```
3. ```push origin```
4. ```checkout -b``` - *talk about it later*
5. ```checkout``` - *talk about it later*
etc etc.

We have a option in git called git alias, means we can create a shortcut lets say of these long git commands.



So, this is the way to create alias now, as we see it is just a configuration that means these alias can be global or local.



***you can create as much alias as you want***
